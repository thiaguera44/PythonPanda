```python
#Novo DF com colunas desejadas...
ca_df = combustiveis_df[['Revenda', 'Municipio','Produto', 'Valor de Venda']]
display(ca_df)
```

```python
print(ca_df.loc[9::19])
```

```python
# Criando DF gas_df contendo apenas 4 colunas
# (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustivel sendo Gasolina

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']
display(gas_df)

display(gas_df['Valor de Venda'].max())
```

```python
# Qual o posto e municipio mais caro do país?

display(gas_df[['Revenda','Municipio', 'Valor de Venda']].max())
```

```python
#DataFrame.loc[] com múltiplas condições para filtragens
#Quais os preços, postos que vendem ETANOL na minha cidade (BELO HORIZONTE)

bh_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df['Municipio'] == 'BELO HORIZONTE')] 
bh_df.sort_values(by = 'Valor de Venda', inplace = True)
display(bh_df)
```

```python
#Qual a média de preços dos combustiveis GASOLINA E GASOLINA ADITIVADA do Bairro BELA VISTA em SÃO PAULO?

display(combustiveis_df.loc[(combustiveis_df['Bairro'] == 'MOOCA') & 
                            (combustiveis_df['Municipio'] == 'SAO PAULO') &
                            (combustiveis_df['Produto'].isin(["GASOLINA", "GASOLINA ADITIVADA"])), 
                            ['Valor de Venda']].mean())
```

```python
# Mostrar média de valor de venda POR COMBUSTIVEL do brasilzão todo?

media_combustivel_df = ca_df[['Produto','Valor de Venda']].groupby(by = 'Produto').mean().round(3)
display(media_combustivel_df)
```

```python
#Quero adicionar uma coluna de valor booleano chamada "Ativo"
# que, por padrão, vai ser True para TODAS as linhas

combustiveis_df['Ativo'] = True 

print(combustiveis_df.shape)
print(combustiveis_df.info())
```
