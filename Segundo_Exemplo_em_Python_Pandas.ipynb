**Segundo exemplo de manipulação de dados usando Python Pandas**

Data: 09  e 10 de fevereiro de 2023



```python
# Importando biblioteca pandas no Python

import pandas as pd

combustiveis_df = pd.read_excel('precos.xlsx')
```

```python
display(combustiveis_df)
```

```python
#Inserção de dados simples

combustiveis_df['Ativo'] = True

display(combustiveis_df.head(5))
```

```python
# Vou colocar uma coluna chamada "OBS"
# que preencha automaticamente "MELHOR CIDADE"
# quando o municipio for BELO HORIZONTE, MG...

combustiveis_df['Obs'] = ['MELHOR CIDADE' if municipio == 'BELO HORIZONTE' else '' for municipio in combustiveis_df['Municipio']]
  
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['BELO HORIZONTE','INDAIATUBA','CONTAGEM']),['Municipio','Obs']])
```

```python
# Preencher uma coluna "Valor de Venda - Satus" que verifica:
# se o valor de venda for maior que 6,50 , ele avisa que está caro

import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Está caro', 'Está barato')

display(combustiveis_df[['Municipio','Produto','Valor de Venda', 'Valor de Venda - Status']].head(10))
```

```python
# Calcular postos de gasolina por abitante na amostragem de combustiveis nov/2021

num_hab_df = pd.read_csv("ibge_num_habitantes_estimado.csv", sep=";")
num_hab_df.rename(columns = {"Estado":"Estado - Sigla"}, inplace=True)

display(num_hab_df)
```

```python
# MERGE dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_hab_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#destruir coluna completamente vazia 

merge_df.dropna(axis=1, inplace = True)
print(merge_df.info())
```

```python
colunas = ['Regiao - Sigla', 'Nome da Rua','Numero Rua', 'Bairro', 
           'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda', 
           'Unidade de Medida', 'Bandeira','Ativo', 'Valor de Venda - Status']

merge_df.drop(labels = colunas, axis = 1, inplace = True)
print(merge_df.info())
```

```python
display(merge_df.head(100))
```

```python
#Removendo linhas duplicadas
merge_df.drop_duplicates(inplace = True)

display(merge_df)
```

```python
#Agrupar e contar quantos postos tem na cidade..

postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Numero de Postos"}, inplace=True)
display(postos_por_municipio_df)
```

```python
postos_por_municipio_df['PostosPorHabitante'] = postos_por_municipio_df['Numero de Postos'] / postos_por_municipio_df['NumHabitantes2021']

display(postos_por_municipio_df)
```
